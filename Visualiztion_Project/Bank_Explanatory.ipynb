# Bank Marketing Campaign Analysis 

## Investigation Overview
In this investigation, I wanted to understand what factors affected the results of the bank's marketing campaign, and by understanding the dataset and getting results; we can use them to optimize the performance of future campaigns.

## Data Overview
The data includes about 11000 observations of the marketing campaign made by the bank to ask clients to subscribe to term deposits. Data about clients include their age, balance, how often they were contacted, duration of contacts and their marital status. It also includes wheter they have housing or personal loans or they have defaults in their credit.

```python
#importing packages that will be implemented in next lines of coding
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
#loading the dataset
camp_data = pd.read_csv('bank_data.csv')
```

```python
# dropping unnecessary columns
camp_data.drop(['pdays', 'previous', 'poutcome', 'day'], axis = 1, inplace = True)
```

## Effect of job on the outcome 
### The analysis of the data showed that Retired and Students were significantly more likely to subscribe to term deposits while jobs like Blue-collar and Services were less likely to subscribe.

```python
plt.figure(figsize = (8,6))
freq = camp_data['job'].value_counts()
order = freq.index
sb.countplot(data = camp_data, x = 'job', hue = 'deposit', order = order)
plt.title('Effect of Job on term deposit')
plt.xlabel('Job')
plt.ylabel('Number of Clients')
plt.xticks(rotation = 90);
```

## Effect of marital status on the outcome
### It was found that Married people are less likely to subscribe

```python
plt.figure(figsize = (8,6))
freq = camp_data['marital'].value_counts()
order = freq.index
sb.countplot(data = camp_data, x = 'marital', hue = 'deposit', order = order)
plt.title('Effect of marital status on deposit subscribes')
plt.xlabel('Marital Status')
plt.ylabel('Number of Clients');
```

## Effect of Housing Loans
### It was found that people who have housing loans are less likely to subscribe

```python
plt.figure(figsize = (8,6))
freq = camp_data['housing'].value_counts()
order = freq.index
sb.countplot(data = camp_data, x = 'housing', hue = 'deposit', order = order)
plt.title('Effect of Housing loans on deposit subscribes')
plt.xlabel('Has Housing Loan?')
plt.ylabel('Number of Clients');
```

## Effect of Device used
### People who use cellular subscribe more often than other devices

```python
plt.figure(figsize = (8,6))
freq = camp_data['contact'].value_counts()
order = freq.index
sb.countplot(data = camp_data, x = 'contact', hue = 'deposit', order = order)
plt.title('Effect of Device used')
plt.xlabel('Contact Device')
plt.ylabel('Number of Clients');
```

## Effect of Age on the outcome
### It was found that people who are younger than 35 or older than 60 are more likely to subscribe

```python
plt.figure(figsize = (10,8))
plt.hist(data = camp_data[camp_data['deposit'] == 'yes'], x = 'age', alpha = 0.5, label = 'Deposit = Yes')
plt.hist(data = camp_data[camp_data['deposit'] == 'no'], x = 'age', alpha = 0.5, label = 'Deposit = No')
plt.title('Effect of Age')
plt.xlabel('Age')
plt.ylabel('Number of Clients')
plt.legend();
```

## Effect of Balance on the outcome 
### It was found that people with higher balance in their account are more likely to subscribe

```python
plt.figure(figsize = (10,8))
plt.hist(data = camp_data[camp_data['deposit'] == 'yes'], x = 'balance', alpha = 0.5, label = 'Deposit = Yes')
plt.hist(data = camp_data[camp_data['deposit'] == 'no'], x = 'balance', alpha = 0.5, label = 'Deposit = No')
plt.title('Effect of Balance in the account')
plt.xlabel('Balance')
plt.ylabel('Number of Clients')
plt.legend();
```

## Effect of Duration on the outcome
### It was found that the more duration a client takes the more likely he/she is to subscribe

```python
plt.figure(figsize = (10,8))
plt.hist(data = camp_data[camp_data['deposit'] == 'yes'], x = 'duration', alpha = 0.5, label = 'Deposit = Yes')
plt.hist(data = camp_data[camp_data['deposit'] == 'no'], x = 'duration', alpha = 0.5, label = 'Deposit = No')
plt.title('Effect of contact duration')
plt.xlabel('Duration')
plt.ylabel('Number of Clients')
plt.legend();
```

## It was noticed that when people are contacted fewer number of times, and their contact duration is high. They are more likely to subscribe

```python
g = sb.FacetGrid(data = camp_data, hue = 'deposit', height = 5)
g.map(plt.scatter,'campaign','duration', alpha = 0.5)
plt.title('Number of contacts vs. Duration of contacts')
plt.xlabel('Number of contacts')
plt.ylabel('Duration of contacts')
plt.legend();
```
